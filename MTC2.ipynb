```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns

from pathlib import Path
```

```python
# create a Path instance and filter for only csv files
files = Path(r"C:\Users\User\Desktop\Data sets").rglob("*.csv")

# read in all the csv files
df= [pd.read_csv(file) for file in files]


# lump into one table

df= pd.concat(df)
df
```

# Checking the null values

```python
df.isnull().sum()
```

# Cheking the total of duplicated values

```python
df.duplicated().sum()
```

# Removing the duplicated values

```python
df.drop_duplicates( inplace=True)
```

# Cheking how long a certain user have been active
### The most extended a user have been active is a year

```python
period=df.groupby(['USER','SDP_STATUS']).size()
period
```

# Saving the period data frame to a csv file

```python
period.to_csv('Extended.csv')
```
